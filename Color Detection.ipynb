```python
import numpy as np
def getLimits(color):
    c=np.uint8([[color]])
    hsvC=cv2.cvtColor(c,cv2.COLOR_BGR2HSV)
    
    """lowerLimit = (hsvC[0][0][0] - 10 if hsvC[0][0][0] - 10 >= 0 else 0, 100, 100)
    upperLimit = (hsvC[0][0][0] + 10 if hsvC[0][0][0] + 10 <= 180 else 180, 255, 255)"""
    
    
    lowerLimit=hsvC[0][0][0]-5,100,100
    upperLimit=hsvC[0][0][0]+5,255,255
    
    """lowerLimit=hsvC[0][0][0]-25,100,100
    upperLimit=hsvC[0][0][0]+25,255,255"""
    
    
    lowerLimit=np.array(lowerLimit,dtype=np.uint8)
    upperLimit=np.array(upperLimit,dtype=np.uint8)
    
    return lowerLimit,upperLimit
```

```python
import cv2
from PIL import Image


cap=cv2.VideoCapture(0)

color=[0,255,255]
while True:
    succ,frame=cap.read()
    if succ:
        hsvFrame=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
        lowerLimit,upperLimit=getLimits(color)
        mask=cv2.inRange(hsvFrame,lowerLimit,upperLimit)
        mask_=Image.fromarray(mask)
        
        bbox=mask_.getbbox()
        
        if bbox is not None:
            x1,y1,x2,y2=bbox
            cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),5)
        cv2.imshow("frame",frame)
        
    
    if cv2.waitKey(1)==ord('q'):
        break
        
cv2.destroyAllWindows()
cap.release()
```
